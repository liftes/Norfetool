```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl


# 创建示例数据
data = np.random.rand(10, 10)

# 创建热力图
fig, ax = plt.subplots()
cax = ax.imshow(data, cmap='viridis')

# 创建 colorbar
cbar = fig.colorbar(cax)

# 使用自定义函数设置 colorbar 刻度线朝内
Dplt.Set_colorbar_ticks_inward(cbar, )
# Dplt.Set_colorbar_ticks_outward(cbar, )

# 显示热力图
plt.show()


plt.figure(figsize=(3,3))
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16],label="test")  # 示例图表
plt.xlabel("test x $x=y$")
Dplt.Set_axis_formatting('both', nbins=6, decimals=0)  # 设置x轴和y轴
plt.legend()
Dplt.SaveFig(2, "test", )
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

data = np.random.rand(10, 20) * 2 - 1  # 在 [-1, 1] 范围内
cmap = plt.cm.coolwarm
fig, ax = plt.subplots(figsize=(4, 2))
img = ax.imshow(data, aspect='auto', cmap=cmap, origin='lower', vmin=-1, vmax=1)

# 添加顶部 colorbar
divider = make_axes_locatable(ax)
cax = divider.append_axes("top", size="10%", pad=0.3)
cbar = fig.colorbar(img, cax=cax, orientation='horizontal')
Dplt.Set_colorbar_ticks_outward(cbar)
```
